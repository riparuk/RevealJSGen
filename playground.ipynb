```python
from openai import OpenAI
client = OpenAI()

response = client.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {
      "role": "system",
      "content": [
        {
          "text": "You are an Intelligent bot for making presentations, user may ask to to create full presentation or edit existing presentation. if user ask for the first time or want to edit the wohole presentation, provide generateSlidesAll(). If user ask for edit presentation, provide generateSlides(). If user ask in advance, provide one or more function to call.",
          "type": "text"
        }
      ]
    },
    {
      "role": "user",
      "content": [
        {
          "text": "Edit slides 5 and 9 to include more information about the product and make it simple",
          "type": "text"
        }
      ]
    }
  ],
  response_format={
    "type": "text"
  },
  tools=[
    {
      "type": "function",
      "function": {
        "name": "generateSlides",
        "strict": True,
        "parameters": {
          "type": "object",
          "required": [
            "index",
            "prompt"
          ],
          "properties": {
            "index": {
              "type": "number",
              "description": "Index of the slide"
            },
            "prompt": {
              "type": "string",
              "description": "Prompt for generating slides"
            }
          },
          "additionalProperties": False
        },
        "description": "Generate slides for specific index"
      }
    },
    {
      "type": "function",
      "function": {
        "name": "generateSlidesAll",
        "strict": True,
        "parameters": {
          "type": "object",
          "required": [
            "prompt"
          ],
          "properties": {
            "prompt": {
              "type": "string",
              "description": "Prompt for generating slides"
            }
          },
          "additionalProperties": False
        },
        "description": "Generate All slides, good for first time"
      }
    }
  ],
  temperature=1,
#   max_completion_tokens=2048,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

```

```python
if response.choices[0].message.tool_calls:
    print(f"Function Call: {response.choices[0].message.tool_calls}")
    for tool_call in response.choices[0].message.tool_calls:
        print(f"Tool Name: {tool_call.function}")
else:
    print("Message", response.choices[0].message.content)
    
print(response.choices[0].message)
```
